```python
def get_stock_metadata():
    stock_metadata = pd.read_csv("yfinance_metadata.csv")  # Load tickers from given dataset
    return stock_metadata["Ticker"].tolist()

# Fetch historical price data
def get_price_data(tickers, start_date="2012-01-01", end_date="2024-12-31"):
    data = yf.download(tickers, start=start_date, end=end_date)["Close"]
    return data.dropna(axis=1, how="any")  # Drop stocks with missing data

# Compute Z-score for mean reversion strategy
def compute_z_scores(price_data, window=20):
    rolling_mean = price_data.rolling(window=window, min_periods=1).mean()
    rolling_std = price_data.rolling(window=window, min_periods=1).std()
    z_scores = (price_data - rolling_mean) / rolling_std
    return z_scores.shift(-1)


# tickers = get_stock_metadata()

# print(f"Fetching data for {len(tickers)} stocks...")
# price_data = get_price_data(tickers)

print("Computing Z-scores...")
z_scores = compute_z_scores(price_data)

print("Filtering stocks with sufficient correlation...")
mean_correlations = price_data.pct_change().corr().mean()
filtered_tickers = mean_correlations[mean_correlations > 0.3].index.tolist()  # Keeping stocks with >0.3 average correlation

print(f"Final universe size: {len(filtered_tickers)}")
final_price_data = price_data[filtered_tickers]
final_z_scores = z_scores[filtered_tickers]

# Save results
final_price_data.to_csv("filtered_price_data.csv")
final_z_scores.to_csv("z_scores.csv")
print("Filtered price data and Z-scores saved.")
```
